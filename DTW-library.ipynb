```python
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw
import numpy as np
import matplotlib.pyplot as plt

# 시퀀스 Q와 C 설정
Q = np.array([3.5 ,4.5 ,3.6 ,4.7]).reshape(-1, 1)
C = np.array([3.4 ,4.6 ,3.5]).reshape(-1, 1)

# fastDTW 실행 및 결과 출력
distance, path = fastdtw(Q, C, dist=euclidean)

print("Distance: ", distance)
print("Path: ", path)

path_x, path_y=zip(*path)

plt.figure(figsize=(10,5))
plt.plot(Q.flatten()[list(path_x)], label='Q')
plt.plot(C.flatten()[list(path_y)], label='C')
plt.legend()
plt.show()
```
